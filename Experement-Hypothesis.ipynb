```python
def fixed_vertex(V):
    
    y_SLS = []
    y_SLP = []
    y_SLH = []
    y_UnS = []
    y_UnP = []
    y_UnH = []
    y_RwS = []
    y_RwP = []
    y_RwH = []
    y_V = []
    y_Pr = []
    for i in range(0, 10):
        print('done')
        r_SLS, r_SLP, r_SLH, r_UnS, r_UnP, r_UnH, r_RwS, r_RwP, r_RwH, r_V, r_Pr = Result(V, i/10)
        
        
        y_SLS.append(r_SLS)
        y_SLP.append(r_SLP) 
        y_SLH.append(r_SLH)
        y_UnS.append(r_UnS)
        y_UnP.append(r_UnP) 
        y_UnH.append(r_UnH)
        y_RwS.append(r_RwS)
        y_RwP.append(r_RwP) 
        y_RwH.append(r_RwH)
        y_V.append(r_V)
        y_Pr.append(r_Pr)
        
    return y_SLS, y_SLP, y_SLH, y_UnS, y_UnP, y_UnH, y_RwS, y_RwP, y_RwH, y_V, y_Pr
    
```

```python
def run_exp(V):
    r_SLS, r_SLP, r_SLH, r_UnS, r_UnP, r_UnH, r_RwS, r_RwP, r_RwH, r_V, r_Pr = fixed_vertex(V)

    x = range(0,10) 
    #plt.plot(x, r_UnH, color ='r', label ='UnH') 
    plt.plot(x, r_RwH, color ='b', label ='RwH') 
    #plt.plot(x, r_SLH, color ='g', label ='SLH') 
    #plt.plot(x, r_UnS, color ='y', label ='UnS') 
    plt.plot(x, r_RwS, color ='c', label ='RwS') 
    #plt.plot(x, r_SLS, color ='m', label ='SLS') 
    plt.plot(x, r_V, color = 'k', label ='V')
    plt.plot(x, r_Pr, color = 'r', label ='Pr')
    plt.xlabel('probability') 
    plt.ylabel('average_conductance') 

    plt.title("Ncut")
    plt.legend(loc='upper right')
    plt.rcParams["figure.figsize"] = [20,9]
    path = os.getcwd()
    plt.savefig(path + '/Img/' + str(V) + '2.png')
    plt.close()
    
    return 
```
